# Introduction

The Aspiring Mind Employment Outcome 2015 (AMEO) dataset, released by Aspiring Minds, provides valuable insights into the employment outcomes of engineering graduates. Focused exclusively on individuals with engineering backgrounds, this dataset offers a comprehensive look at factors influencing salary, job titles, and job locations. Moreover, it encompasses standardized scores across three key domains: cognitive skills, technical expertise, and personality traits. With approximately 40 independent variables and 4000 data points, the dataset presents a rich source of information, blending both continuous and categorical data. Additionally, demographic features are included, ensuring a holistic understanding of the candidate pool. Each candidate is uniquely identified within the dataset, enhancing traceability and analysis capabilities

# Objective

The primary objective of this Exploratory Data Analysis (EDA) is to conduct a comprehensive examination of the given dataset, focusing specifically on elucidating the connections between different variables and the target variable, Salary.

This analysis aims to achieve the following objectives:

1. Thoroughly describing the features present in the dataset.
2. Identifying any discernible patterns or trends within the data.
3. Exploring the relationships between the independent variables and the target variable (salary).
4. Detecting any outliers or anomalies that may be present in the dataset.
5. Providing practical insights and recommendations based on the findings of the analysis.
6. Assessing the data quality and completeness to ensure reliability of conclusions drawn.

# Datatypes Conversion

### DOL - Date of Leaving.

The 2015 survey implies an assumption: respondents marking 'present' for Date of Leaving (DOL) likely left within the same year. Substituting 'present' in DOL with '2024-02-17' reflects this assumption. DOJ and DOL are then converted to datetime for analysis.

```python
df['DOL'].replace('present','2015-12-31', inplace = True)
df['DOL'] = pd.to_datetime(df['DOL'])
df['DOJ'] = pd.to_datetime(df['DOJ'])
df.head()
```

```python
categorical = ['Designation','JobCity', 'Gender','10board','12board','CollegeTier','Degree',
              'Specialization','CollegeCityTier','CollegeState']
for cat in categorical:
    df[cat] = df[cat].astype('category')
```

```python
df.dtypes
```

```python
delta = (df['DOL'] - df['DOJ'])
tenure = np.zeros(len(df))
for i, date in enumerate(delta):
    tenure[i] = round(date.days/365,2)
df['Tenure'] = tenure
```

```python
def titlecase_column(column_name):
    df[column_name] = df[column_name].astype(str).str.title()
categorical_columns = df.select_dtypes(include='object').columns
for column in categorical_columns:
    titlecase_column(column)
```

```python
df = df.drop(columns = ['Unnamed: 0', 'ID','CollegeID','CollegeCityID'])
df.head()
dates = df[(df['DOL'] < df['DOJ'])].shape[0]
print(f'DOL is earlier than DOJ for {dates} observations.')
print(df.shape)
```

```python
df = df.drop(df[~(df['DOL'] > df['DOJ'])].index)
print(df.shape)
```

```python
print((df['10percentage'] <=10).sum())
print((df['12percentage'] <=10).sum())
print((df['collegeGPA'] <=10).sum())
```

```python
df.loc[df['collegeGPA']<=10,'collegeGPA'].index
df.loc[df['collegeGPA']<=10,'collegeGPA'] = (df.loc[df['collegeGPA']<=10,'collegeGPA']/10)*100
df.head()
```

```python
print((df==0).sum()[(df==0).sum() > 0])
```

```python
(df==-1).sum()[(df==-1).sum()>0]/len(df)*100
```

```python
df = df.drop(columns = ['MechanicalEngg', 'ElectricalEngg', 'TelecomEngg', 'CivilEngg'])
```

```python
df['10board'] = df['10board'].astype(str)
df['12board'] = df['12board'].astype(str)
df['JobCity'] = df['JobCity'].astype(str)
```

```python
df['10board'] = df['10board'].replace({'0': np.nan})
df['12board'] = df['12board'].replace({'0': np.nan})
df['GraduationYear'] = df['GraduationYear'].replace({0: np.nan})
df['JobCity'] = df['JobCity'].replace({'-1': np.nan})
df['Domain'] = df['Domain'].replace({-1: np.nan})
df['ElectronicsAndSemicon'] = df['ElectronicsAndSemicon'].replace({-1: 0})
df['ComputerScience'] = df['ComputerScience'].replace({-1: 0})
df['ComputerProgramming'] = df['ComputerProgramming'].replace({-1: np.nan})

df['10board'] = df['10board'].astype('category')
df['12board'] = df['12board'].astype('category')
df['JobCity'] = df['JobCity'].astype('category')
```

```python
df['10board'].fillna(df['10board'].mode()[0], inplace=True)
df['12board'].fillna(df['12board'].mode()[0], inplace=True)
df['GraduationYear'].fillna(df['GraduationYear'].mode()[0], inplace=True)
df['JobCity'].fillna(df['JobCity'].mode()[0], inplace=True)

df['Domain'].fillna(df['Domain'].median(), inplace=True)
df['ComputerProgramming'].fillna(df['ComputerProgramming'].median(), inplace=True)
```

```python
def correct_string_data(data):
    df[data] = df[data].str.lower().str.strip()
```

```python
textual_columns = ['Designation','JobCity','10board','12board','Specialization','CollegeState']
for col in textual_columns:
    correct_string_data(col)
for col in textual_columns:
    print(f'Number of unique values in {col} without inconsistency : {df[col].nunique()}')
```

```python
def collapsing_categories(df, data):
    for category in df[data].unique():
        min_count = df[data].value_counts()[:10].min()
        if df[df[data] == category][data].value_counts()[0] < min_count:
            df.loc[df[data] == category, data] = 'other'
```

```python
for cols in textual_columns:
    collapsing_categories(df, cols)
```

### Univariate Analysis - Statistical Non Visual Analysis

Exploratory Data Analysis (EDA) serves as a fundamental cornerstone in data analysis, utilizing a blend of statistical methods and visual exploration to unveil hidden patterns and glean insights from datasets. By leveraging descriptive statistics, insightful visualization, and meticulous preprocessing techniques, EDA not only ensures the integrity of the data but also pinpoints outliers and anomalies. Through iterative exploration and meticulous documentation of significant discoveries, EDA lays a solid groundwork for informed, data-driven decision-making processes.

```python
discrete_df = df.select_dtypes(include=['object'])
numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
sns.barplot(x='Salary', y='Designation',
            data=df3,
            capsize=0.1,
            width=0.3,
            ax=ax)
ax.axvline(df3['Salary'].mean(), color='k',
           linestyle=':',
           linewidth=2, label='Overall\nAvg. Salary')
ax.set_title('Avg Salary for Each Designation after pursuing Computer Science Engineering')
ax.legend(loc='upper right', bbox_to_anchor=(1.4, 1))
ax.set_xlabel('')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.tight_layout()
plt.show()
```
